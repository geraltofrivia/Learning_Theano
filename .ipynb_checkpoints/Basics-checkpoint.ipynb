## Shared Variable

```python
#Shared variables being updated at each function call
state = theano.shared(0)
inc = T.iscalar('a') #The type of this tensor must be same as that of the one defined in the shared function

#This function will update the value after giving the answer. Bookkeeping happens after
accumulator = theano.function([inc],state,updates=[(state, state**2+inc)])
```

## Eg: Togglable shared variable
A function that toggles the value of a shared variable (a matrix) whenever it runs. 

First off, we need to have a matrix as a shared variable.
The problem is, that shared variables can't be defined as easy as:

    `s = theano.shared([[1,2],[2,1]])`
    
We need another way to initialize an empty matrix in the params. 
The best way to do that, is to use zeros function of numpy

```python
shared_matrix = theano.shared(numpy.zeros((2,2)))
#This will create an empty 2x2 matrix and the shared variable will be created according to that.

#Let's set an initial value in the variable
shared_matrix.set_value([[1,2],[2,1]])

#Now let's create a function that will invert this matrix. Since invert is reflexive, this simply becomes a toggle function
toggle = theano.function([],shared_matrix,updates=[(shared_matrix, theano.tensor.nlinalg.matrix_inverse(shared_matrix))])
```

```python
#Let's test this function. Simply execute this every now and then!
print toggle()
```

## Random Variables/Random Stream
Not so straightforward with Numpy functions and variables. Since the variables and functions are first expressed symbolicaly and then given a value, getting randomness there isn't so straightforward. To get there, to have a no input function generating random variables, we use "Random Streams"

Random streams too are shared variables/act like it

```python
#Let's generate and plot matrices generated randomly
random_streams = theano.tensor.shared_randomstreams.RandomStreams(seed=255)
rv_uni = random_streams.uniform((2,2))
rv_nor = random_streams.normal((2,2))
realize_random_uni_matrix = theano.function([],rv_uni)
realize_random_nor_matrix = theano.function([],rv_nor)
```

```python
#Use the function to generate a shitload of these and store their frequencies to check if they really perform randomly
def flatten_matrix(_mat, _decimals):
    ''' Whatever be the matrix' dimension, reshape it into a 1D vector and return rounded values'''
    #First convert the matrix into a numpy matrix
    np_arr = numpy.array(_mat)

    #Then flatten the matrix
    np_arr = np_arr.flatten()
    
    #Use this logic to round off the elements
    np_arr = [ int(x*_decimals)/float(_decimals) for x in np_arr ]
    
    #Finally convert it into a 1D array and return
    return list(np_arr)

def plot_output_frequency(_function, _detail = 1000):
    
    #Generate a list of hashmap of value, frequency for each variable
    test_mat = numpy.array(realize_random_nor_matrix()).shape
    data_frequency = []
    for v in range(test_mat[0]*test_mat[1]):
        data_frequency.append({})
    
    #Count the data frequency
    for i in range(100000):
        random_mat = _function()
        random_vec_normalized = flatten_matrix(random_mat, _detail)
        #For every element of the vector, find a place in dictionary. if not there, create it
        for index in range(len(random_vec_normalized)):
            try:
                data_frequency[index][random_vec_normalized[index]] += 1
            except KeyError:
                data_frequency[index][random_vec_normalized[index]] = 1
                
    
    #Plot the data frequency thingy
    for frequency_dict in data_frequency:
        X = numpy.array(frequency_dict.keys())
        Y = numpy.array([ frequency_dict[key] for key in X])
        plt.scatter(X,Y)
        plt.show()
        print "mean= ", Y.mean(), "\n std= ", Y.std()
        
%matplotlib inline
print "Plot of Uniform Streams\n"
plot_output_frequency(realize_random_uni_matrix)
print "Plot of Uniform Streams\n"
plot_output_frequency(realize_random_nor_matrix)
```
